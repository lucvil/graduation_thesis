## 初期設定

```python
import json
import sys
import time
import random
import numpy as np
import statistics
import matplotlib.pyplot as plt
import japanize_matplotlib
import pandas as pd

def average_record(record):
	record_list = []
	for record_item in record.values():
		record_list.extend(record_item)
	return statistics.mean(record_list)

def average_system_record(record, want_item):
	record_list = []
	for record_item in record.values():
		record_list.extend(record_item[want_item])
	return statistics.mean(record_list)

```
